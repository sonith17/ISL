#Exercise 1: Create histograms, KDE, and ECDF plots to explore distributions in the tips dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the tips dataset
tips = sns.load_dataset('tips')
sns.histplot(data=tips, x='total_bill', kde=True)
plt.title('Histogram with KDE for Total Bill')
plt.show()
```

```python
sns.histplot(data=tips, x='tip', hue='sex',multiple='stack')
plt.title('Total Bill Distribution by Sex')
plt.legend(title='Sex')
plt.xlabel('Total Bill ($)')
plt.legend(labels=['Female', 'Male'])
plt.ylabel('Count')
plt.show()
```

```python
sns.ecdfplot(data=tips, x='tip')
plt.title('ECDF for Total Bill')
plt.show()
```

# Exercise 2: Use box plots, violin plots, and count plots to analyze categorical relationships.

```python
sns.boxplot(data=tips, x='day', y='total_bill',hue='time',palette='Set2')
plt.title('Box Plot of Total Bill by Day')
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.show()
```

```python
iris = sns.load_dataset('iris')
sns.violinplot(data=iris, x='species', y='petal_length')
plt.title('Violin Plot of Petal Length by Species')
plt.xlabel('Species')
plt.ylabel('Petal Length')
```

```python
# Plot a count ofdayvalues intips, grouped bysmokerusinghue.
sns.countplot(data=tips, x='day', hue='smoker')
plt.title('Count of Days by Smoker')
plt.xlabel('Day')
plt.ylabel('Count')
plt.legend(title='Smoker')
plt.show()
```

# Exercise 3: Visualize relationships between numerical variables and add regression lines.

```python
# Create a scatter plot oftotal_billvs.tipintips, colored bydaywith theSet2palette.
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day', palette='Set2')
plt.title('Scatter Plot of Total Bill vs. Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
```

```python
#Uselmplot()to add regression lines fortotal_billvs.tip, split bysmoker.
sns.lmplot(data=tips, x='total_bill', y='tip', hue='smoker')
plt.title('Regression Plot of Total Bill vs. Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

```python
# For thediamondsdataset, plotcaratvs.priceand annotate regions for "expensive" and "affordable" diamonds.
diamonds = sns.load_dataset('diamonds')
sns.scatterplot(data=diamonds, x='carat', y='price')
plt.title('Scatter Plot of Carat vs. Price')
plt.xlabel('Carat')
plt.ylabel('Price')
plt.text(1.5, 15000, 'Expensive', fontsize=12, color='red')
plt.text(0.5, 2000, 'Affordable', fontsize=12, color='green')
plt.show()
```

# Exercise 4: Explore correlations and multi-variable relationships.

```python
# Compute and visualize the correlation matrix for numerical columns intipsas a heatmap.
corr_matrix = tips.corr(numeric_only=True)
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
# Generate a pair plot for theirisdataset, colored byspecieswith KDE diagonals.
sns.pairplot(data=iris, hue='species', diag_kind='kde')
plt.suptitle('Pair Plot of Iris Dataset')
plt.show()
```

```python
# Create a joint plot forsepal_lengthvs.sepal_widthiniris.
sns.jointplot(data=iris, x='sepal_length', y='sepal_width')
plt.show()
```

# Exercise 5: Use faceting to compare subsets of data.

```python
# UseFacetGridto create histograms oftotal_billsplit bydayintips.
g = sns.FacetGrid(tips, col='day')
g.map(sns.histplot, 'total_bill')
plt.show()
```

```python
# Generate a displot oftotal_billwithcol='time'andhue='sex', using stacked histograms.
sns.displot(data=tips, x='total_bill', col='time', hue='sex', multiple='stack')
```

# Exercise 6: Visualize distributions with non-standard plots using the penguins dataset.

```python
# Drop missing values inpenguinsand create a swarm plot of flipper_length_mm by species.
penguins = sns.load_dataset('penguins').dropna()
sns.swarmplot(data=penguins, x='species', y='flipper_length_mm')
plt.title('Swarm Plot of Flipper Length by Species')
plt.xlabel('Species')
plt.ylabel
```

```python
# Generate a boxen plot ofbody_mass_gbyspecies.
sns.boxenplot(data=penguins, x='species', y='body_mass_g')
plt.title('Boxen Plot of Body Mass by Species')
plt.xlabel('Species')
plt.ylabel('Body Mass (g)')
plt.show()
```
